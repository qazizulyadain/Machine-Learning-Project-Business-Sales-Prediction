### Data Cleaning

```python
data.columns
```

```python
drop_columns = ['Product ID', 'brand', 'url', 'name', 'description', 'currency', 'terms', 'Product Category']
data_clean = data.drop(columns=drop_columns)
```

```python
data_clean.sample(10)
```

```python
data_clean.isnull().sum()
```

```python
data_clean.dtypes
```

```python
## IQR
## Q3 - Q1

Q1 = data_clean['Sales Volume'].quantile(0.25)
Q3 = data_clean['Sales Volume'].quantile(0.75)

IQR = Q3 - Q1

data_clean = data_clean[(data_clean['Sales Volume'] >= Q1 - 1.5 * IQR) & (data_clean['Sales Volume'] <= Q3 + 1.5 * IQR)]
```

```python
data_clean.sample(10)
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Promotion', y='Sales Volume', data=data_clean)
plt.title("Sales Volume By Promotions")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder


data_clean['Promotion'] = data_clean['Promotion'].map({"Yes": 1, "No": 0})
data_clean['Seasonal'] = data_clean['Seasonal'].map({"Yes": 1, "No": 0})
```

```python
data_clean.columns
```

```python
data_clean['season'].unique()
```

```python
data_clean = pd.get_dummies(data_clean, columns=['Product Position', 'section', 'material', 'origin'], drop_first=True)

season_mapping = {"Spring": 1, "Summer": 2, "Autumn": 3, "Winter": 4}
data_clean['season'] = data_clean['season'].map(season_mapping)
```

```python
data_clean.sample(10)
```

```python
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression
from sklearn.ensemble import RandomForestRegressor

data_clean['price_categories'] = pd.cut(data_clean['price'], bins=[0, 20, 50, 100, np.inf], labels=['Low', "Medium", "High", "Very Hight"])
```

```python
data_clean = pd.get_dummies(data_clean, columns=['price_categories'], drop_first=True)
```

```python
X = data_clean.drop("Sales Volume", axis=1)
y = data_clean['Sales Volume']
```

```python
corr_with_target = X.corrwith(y).abs().sort_values(ascending=False)
corr_with_target
```

```python
select_corr = corr_with_target[corr_with_target > 0.1].index
```

```python
select_corr
```

```python
var_thresh = VarianceThreshold(threshold=0.01)
X_var = var_thresh.fit_transform(X)
selected_var = X.columns[var_thresh.get_support()]
```

```python
selected_var
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
importances
```

```python
selected_imp = importances.head(10).index
```

```python
selected_imp
```

```python
selected_features = list(set(selected_imp) | set(select_corr))
```

```python
X_selected = X[selected_features]
```

```python
selected_features
```

```python
X_selected
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
